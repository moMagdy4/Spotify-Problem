```python
X_train =  train.drop(columns = ['genre_label'],axis=1)
y_train = train.genre_label
X_test = test
```

```python
x_ones=np.append(np.ones((X_train.shape[0],1)),X_train,axis=1)
```

```python
y_train = y_train.to_numpy()
y_train = y_train.reshape((y_train.shape[0],1))
```

```python
def sigmoid(Theta,X):
  input=np.dot(X,Theta.T)
  output = 1 / (1 + np.exp(-input))
  return output
```

```python
def costFunction(theta, X, y):
   m = y.size  
   j=0
   h=sigmoid(theta,X)
   j =  1/m * np.sum(-y*np.log(h) - (1-y) * np.log(1-h))
   return j
```

```python
def gradientDecsent(iterations, alpha, x, y,theta):
  m=x.shape[0]
  print("shape of y",y.shape)
  print("shape of x",x.shape)
  print("shape of theta",theta.shape)
  cost_history = []
  for i in range(iterations):
    theta = theta - (alpha / m) * np.transpose(sigmoid(theta,x) - y).dot(x)
    cost_history.append(costFunction(theta, x, y))
  return theta, cost_history
```

```python
theta=np.zeros((1,x_ones.shape[1]))
theta, cost_history = gradientDecsent(290000, 0.0002,x_ones, y_train,theta)
print(theta)
```

```python
x_ones_test=np.append(np.ones((X_test.shape[0],1)),X_test,axis=1)
y_pred = sigmoid(theta,x_ones_test)
```

```python
y_pred_final = []
for i in y_pred:
  if i >= 0.5:
    y_pred_final.append(1)
  else:
    y_pred_final.append(0)
print(y_pred_final)      
```

```python
subm={'ID':ID,'genre_label':y_pred_final}
subm=pd.DataFrame(subm)
subm.to_csv("subm.csv",index=False)
```
